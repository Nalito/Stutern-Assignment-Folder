##Question 1

Given the following dataset:

a = np.random.normal(loc=64, scale=2, size=500)
b = np.random.normal(loc=70, scale=2, size=100)

Write the appropriate to display multiple histogram on the same axis. The plots for both datasets show be fully
visible.
[Hint: Use the 'alpha' argument]


```python
a = np.random.normal(loc=64, scale=2, size=500)
b = np.random.normal(loc=70, scale=2, size=100)
```

```python
plt.hist(a, alpha=0.5, color='blue')
plt.hist(b, alpha=0.5, color='red')
plt.legend(["Dataset A", "Dataset B"])
plt.title("Histogram plot showing Dataset A and Dataset B")
plt.show()
```

# Question 3

How would you use matplotlib module to make two different subplots on different axes with the following 
characteristics?

a) First subplot: 
i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p.

b) Second subplot:
i. Create a bar-chart with the 'region' column. 
ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.

In order to make two different subplots on different axes, the first step is to declare the subplot axis for the first subplot. In my case, I placed the first subplot in the axis (2, 2, 1). After that, I placed all my specifications for the first subplot directly under the subplot declaration.
For the second subplot, I repeated the same process as I did for the first. The only thing that changed was that I changed the plot axis to (2, 2, 2).

```python
plt.subplot(2,2,1)
plt.title("Pie Chart of the Sex column")
plt.pie(df["sex"].value_counts(), autopct='%0.2f', labels=["Male", "Female"])
plt.subplot(2,2,2)
plt.bar(df.region.unique(), df.region.value_counts())
plt.xticks(rotation = 45)
plt.xlabel("Region")
plt.ylabel("Counts")
plt.title("Bar plot for Region")
plt.subplots_adjust(wspace=1, hspace=1, bottom=0.1, top=0.9)
plt.show()
```
